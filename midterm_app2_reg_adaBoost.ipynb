```python
import pandas as pd                  # Pandas
import numpy as np                   # Numpy
from matplotlib import pyplot as plt # Matplotlib

# Package to implement ML Algorithms
import sklearn

from sklearn.ensemble import AdaBoostRegressor # Random Forest

# Package for data partitioning
from sklearn.model_selection import train_test_split


# Import packages to implement Stratified K-fold CV
from sklearn.model_selection import KFold # For creating folds

# Import Package to implement GridSearch CV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,r2_score, mean_squared_error
# Importing package for Randomized Search CV
from sklearn.model_selection import RandomizedSearchCV

# Package to record time
import time

# Package for Data pretty printer
from pprint import pprint

# Module to save and load Python objects to and from files
import pickle 

# Ignore Deprecation Warnings
import warnings
warnings.filterwarnings('ignore')

# Display inline plots as vector-based (svg)
%config InlineBackend.figure_formats = ['svg']

%matplotlib inline
```

```python
#Preprocessing data

df_traffic_volume = pd.read_csv('Traffic_Volume.csv')
df_traffic_volume.head()
```

```python
#converting NaN to "None" string
df_traffic_volume['holiday'] = df_traffic_volume['holiday'].fillna('None')
df_traffic_volume['holiday']
```

```python
#checking missing values
df_traffic_volume.isna().sum()
```

```python
#drop unuse column
df_traffic_volume.drop('weather_description',axis=1,inplace=True)
```

```python
#extract datetime variables from the date_time column
df_traffic_volume['date_time'] = pd.to_datetime(df_traffic_volume['date_time'])
df_traffic_volume['month'] = df_traffic_volume['date_time'].dt.month_name()
df_traffic_volume['day'] = df_traffic_volume['date_time'].dt.day_name()
df_traffic_volume['time'] = df_traffic_volume['date_time'].dt.hour
df_traffic_volume.drop('date_time',axis=1,inplace=True)
```

```python
#check missing values
df_traffic_volume.isna().sum()
```

```python
#create predictors and response variable
X_traffic= df_traffic_volume.drop(columns = ['traffic_volume'])
y_traffic = df_traffic_volume['traffic_volume']
```

```python
#transform categorical predictors into dummies
X_traffic_cate_var = ['holiday','weather_main','month','day','time']
X_traffic_encoded = pd.get_dummies(X_traffic, columns = X_traffic_cate_var)
```

```python
X_traffic_encoded
```

```python
#partition data into train/test dataset
#create cv folds
train_X_traffic, test_X_traffic, train_y_traffic, test_y_traffic = train_test_split(X_traffic_encoded, y_traffic, test_size = 0.2, random_state = 1)
folds_3 = KFold(n_splits = 3, shuffle = True, random_state = 100)
```

```python
adaBoost_reg_mod = AdaBoostRegressor(random_state=42)
```

```python
n_estimators = [int(x) for x in np.linspace(start = 90, stop = 150, num = 30)]
learning_rate = [.05,.1,.15,.2,.25,.3,.35,.4,.45,.5]

hyperparameters_rand = {
    'n_estimators': n_estimators,
    'learning_rate': learning_rate,
    'loss': ['linear', 'square', 'exponential']
}

pprint(hyperparameters_rand)
```

```python
#perform randCV
adaBoost_reg_tune = RandomizedSearchCV(estimator = adaBoost_reg_mod, 
                              param_distributions = hyperparameters_rand,
                              n_iter = 100,
                              scoring = 'r2', 
                              cv = folds_3, 
                              verbose = 1,
                              random_state = 42,
                              n_jobs = -1) # Will utilize all available CPUs 
```

```python
adaBoost_reg_tune.fit(train_X_traffic,train_y_traffic)
```

```python
# Return set of parameters with the best performance
print(adaBoost_reg_tune.best_params_)

# Return the performance metric score
print(adaBoost_reg_tune.best_score_)
```

```python
y_pred_test_adaBoost = adaBoost_reg_tune.predict(test_X_traffic)
r2_score(test_y_traffic,y_pred_test_adaBoost)
```

```python
mean_squared_error(test_y_traffic, y_pred_test_adaBoost, squared=False)
```

```python
bestRegAdaBoost= adaBoost_reg_tune.best_estimator_
```

```python
midterm_pickle_reg_AdaBoost = open('reg_AdaBoost_fixed.pickle','wb')
pickle.dump(bestRegAdaBoost, midterm_pickle_reg_AdaBoost)
midterm_pickle_reg_AdaBoost.close()
```

```python
importance = bestRegAdaBoost.feature_importances_ #gini index purity reduction

feature_imp_traffic_reg_AdaBoost = pd.DataFrame(list(zip(train_X_traffic.columns, importance)),
               columns = ['Feature', 'Importance'])

feature_imp_traffic_reg_AdaBoost = feature_imp_traffic_reg_AdaBoost.sort_values('Importance', ascending = False).reset_index(drop = True)

feature_imp_traffic_reg_AdaBoost
```

```python
# Selecting features with non-zero importance values
feature_imp_nonzero_traffic_reg_AdaBoost = feature_imp_traffic_reg_AdaBoost[feature_imp_traffic_reg_AdaBoost['Importance'] != 0.0]

# Bar plot
plt.figure(figsize=(10, 5), dpi = 100)
plt.barh(feature_imp_nonzero_traffic_reg_AdaBoost['Feature'], feature_imp_nonzero_traffic_reg_AdaBoost['Importance'], color = ['blue', 'red'])

plt.xlabel("Importance")
plt.ylabel("Input Feature")
plt.title("Feature Importance for AdaBoost Model");
```
